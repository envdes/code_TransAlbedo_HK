# Export baseline results
- This script is used to export metric results on very hot days and hot nights.
- Simulation: CNTL

```python
import xarray as xr
import numpy as np
import pandas as pd
import os
home_path = '/gws/ssde/j25a/duicv/yuansun/'
project_path = f'{home_path}0_wrf-cstm_GM-HK/'
night_threshold = 28
day_threshold = 33
var_list = ['TSA', 'TSA_U', 'TSA_R']
region='HK'
```

```python
results = []
for scenario in ['cntl']: #, 'tran_albe'
    temp_file = f'../temperature_trend/data_for_figure/{region}_tsa_{scenario}_urban_rural.csv'
    df_temp = pd.read_csv(temp_file)
    df_temp['time'] = pd.to_datetime(df_temp['time'])
    df_temp['date'] = df_temp['time'].dt.date
    vhds_list = []
    hns_list = []
    for var in var_list:
        df_temp_var = df_temp[['date', var]].copy()
        df_temp_var_max = df_temp_var.groupby('date')[var].max().reset_index()
        df_temp_var_min = df_temp_var.groupby('date')[var].min().reset_index()
        df_vhds = df_temp_var_max[df_temp_var_max[var] >= day_threshold].copy()
        df_hns = df_temp_var_min[df_temp_var_min[var] >= night_threshold].copy()
        df_vhds.rename(columns={var: 'values'}, inplace=True)
        df_hns.rename(columns={var: 'values'}, inplace=True)
        df_vhds['variable'] = var
        df_hns['variable'] = var
        vhds_list.append(df_vhds)
        hns_list.append(df_hns)
    df_vhds_all = pd.concat(vhds_list)
    df_vhds_all['senario'] = scenario
    df_vhds_all['metric'] = 'vhds'
    df_hns_all = pd.concat(hns_list)
    df_hns_all['senario'] = scenario
    df_hns_all['metric'] = 'hns'
    results.append((df_vhds_all, df_hns_all))
df_results = pd.concat([pd.concat(pair) for pair in results], ignore_index=True)
df_results.to_csv('./data_for_figure/extreme_heat_metrics.csv', index=False)
df_results
```
